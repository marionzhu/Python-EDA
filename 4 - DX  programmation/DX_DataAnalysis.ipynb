```python
def process(file_name):
    file_content =  pd.read_csv(f'{file_name}.csv') # load data
    print(file_content.info()) # check basic infos, like data type, number of rows and columns etc.
    print('\n')
    print(file_content.isna().sum()) # check missing values
    print('\n')
    print(file_content.duplicated().sum()) # check duplications
    print('\n')
    print(file_content.describe())

    return file_content
```

### 2, Find payer
Ce client est « payeur » ou non. Il est payeur si 
* Soit pour tous ses contrats « actif » il a payé l’ensemble de ses factures « abonnement »
* Soit son compte client dans son ensemble est créditeur (il a payé à l’entreprise plus qu’il devait payer)!

We start by find the clients who meet the fisrt condition

```python
# merge table and remove unrelated columns
merged = crm.merge(contract, how = 'left',left_on = 'customer_id', right_on = 'crm_id' )\
    .drop(columns = ['_msdyn_segmentid_value','crm_id','crm_estunpim']).rename(columns = {'id':'contract_id'})\
.merge(invoice, how = 'left', on= 'contract_id')\
    .drop(columns = ['address1_line2','date']).rename(columns = {'id':'invoice_id'})\
.merge(ledger, how='left', on = 'invoice_id')\
    .drop(columns = ['contract_id_y', 'amount_y','crm_activitsprincipales'])

merged.head()
```

```python
# find contract is active and type is abonnement customers
actif_ab = merged[(merged['product_id']=='abonnement') & (merged['status'] == 'actif')]\
    .groupby(['invoice_id','customer_id','name'])[['sales', 'profit']].sum().reset_index()

# find if the client paid for each invoice
actif_ab['credit'] = actif_ab['profit'] - actif_ab['sales']

# if any of the credit is less than 0, then we consider the clients doesn't pay "l’ensemble de ses factures « abonnement »"
actif_ab['bad_invoice'] = actif_ab['credit'] < 0
paid_customers = actif_ab.groupby('customer_id')['bad_invoice'].sum().reset_index().query('bad_invoice == 0')

# find paid customers ID
paid_customers_id = paid_customers['customer_id']
paid_customers_id
```

Find the custmers who meet the second conditions and concat it with first group

```python
# find customer's total sales and profit 
cust_ledger = merged.groupby('customer_id')[['sales', 'profit']].sum().reset_index()
cust_ledger['credit'] = cust_ledger['profit']-cust_ledger['sales']

# find customers whose account has credit
credit_customers = cust_ledger[cust_ledger['credit']>=0]

# find customers ID whose account has credit
credit_customers_id = credit_customers['customer_id']
credit_customers_id
```

```python
# merge two parts and find total paid customers
total_paid_customer = pd.concat([credit_customers_id,paid_customers_id]).unique()
total_paid_customer
```

### 3, Find new clients
* Ce client est-il « nouveau » c’est-à-dire que l’ensemble de ses contrats « actif » ont été créés dans l’année en cours!

```python
# find all actif contracts of each customer and the creation_date
contract_creation = merged[merged['status']=='actif']\
    .groupby(['customer_id', 'creation_date']).size().reset_index(name = 'num_contracts')

# print all unique creation date 
print(sorted(contract_creation['creation_date'].unique(), reverse = True))
```

<div style='color:red;'> 
Attention :  potential data pipline errors  
</div>
According to the data, a lot of contracts have already been created in the future, like in the year of 2025 or on Mai, June, July ... of 2024, which doesn't make sense in real life. These could be a potential error of data pipline. <br>
In the excercise, we will just consider all the contracts created <strong>in the year of 2024 or after 2024 </strong> as new clients

```python
# find new customers
first_contract = contract_creation.groupby('customer_id')['creation_date'].min().reset_index(name = 'first_contract')
new_cust = first_contract[first_contract['first_contract']>='2024-01-01']
new_cust_id = new_cust['customer_id'].unique()
new_cust_id
```

### 4, Final result
* customer name
* customer ID
* payer or not 
* new cusomer or not 

**Due to missing values, the result may have a bias**

```python
# final customer table with all infomation needed
crm['payer'] = crm['customer_id'].apply(lambda x: True if x in total_paid_customer else False)
crm['new_customer'] = crm['customer_id'].apply(lambda x : True if x in new_cust_id else False)
crm = crm.drop(columns = '_msdyn_segmentid_value')
crm
```
